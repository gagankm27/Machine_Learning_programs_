# NumPy and Pandas Tutorial

This notebook covers the basics of **NumPy** and **Pandas** with examples.

## NumPy Basics

```
import numpy as np
```

```

# Creating arrays
arr = np.array([1, 2, 3, 4, 5])
arr
```

```

# Array properties
arr.shape, arr.size, arr.dtype
```

```

# Mathematical operations
arr + 10
```

```

# Reshaping arrays
arr2 = np.arange(1, 7)
arr2.reshape(2, 3)
```

## Pandas Basics

```
import pandas as pd
```

```

# Creating a DataFrame
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "Salary": [50000, 60000, 70000]
}

df = pd.DataFrame(data)
df
```

```

# DataFrame information
df.info()
```

```

# Selecting columns
df["Age"]
```

```

# Filtering data
df[df["Age"] > 28]
```

```

# Adding a new column
df["Bonus"] = df["Salary"] * 0.1
df
```

```

# GroupBy example
df.groupby("Age")["Salary"].mean()
```
