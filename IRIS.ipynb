# Importing the Library and Dataset

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
```

# Loading DataSet and Splitting into Target and Variable

```python
iris = load_iris()
x = iris.data
y = iris.target
```

```python
x
```

```python
y
```

# Splitting Dataset into Training and Testing Data 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

# Standardaize Independent Feature in variables by Standard Scalar

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

# Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
pred = model.predict(X_test)
print("Logistic Regression Accuracy:", accuracy_score(y_test, pred))
```

# DecisonTree Classifier Model

```python
from sklearn.tree import DecisionTreeClassifier
model_1 = DecisionTreeClassifier(max_depth=2, random_state=0)
model_1.fit(X_train,y_train)
```

```python
pred_1 = model_1.predict(X_test)
print("Decision Tree Accuracy:", accuracy_score(y_test, pred_1))
```

# Random Forest Classifier Model

```python
from sklearn.ensemble import RandomForestClassifier
model_2 = RandomForestClassifier(n_estimators=100, random_state=42)
model_2.fit(X_train, y_train)
```

```python
pred_2 = model_2.predict(X_test)
print("Random Forest Accuracy:", accuracy_score(y_test, pred_2))
```

# Naive Bayes Model

```python
from sklearn.naive_bayes import GaussianNB
model_3 = GaussianNB()
model_3.fit(X_train, y_train)
```

```python
pred_3 = model_3.predict(X_test)
print("Gaussian NB Accuracy:", accuracy_score(y_test, pred_3))
```

# SVM Model

```python
from sklearn.svm import SVC
model_4 = SVC(kernel='rbf')
model_4.fit(X_train, y_train)
```

```python
pred_4 = model_4.predict(X_test)
print("SVM Accuracy:", accuracy_score(y_test, pred_4))
```

# KNN Model

```python
from sklearn.neighbors import KNeighborsClassifier
model_5 = KNeighborsClassifier(n_neighbors=5)
model_5.fit(X_train, y_train)
```

```python
pred_5 = model_5.predict(X_test)
print("KNN Accuracy:", accuracy_score(y_test, pred_5))
```
