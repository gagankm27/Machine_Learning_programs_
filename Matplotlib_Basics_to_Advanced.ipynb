This notebook teaches Matplotlib step-by-step with explanations and runnable examples.

## 1. Import Libraries

```
import matplotlib.pyplot as plt
import numpy as np
```

## 2. First Line Plot

```

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)
plt.show()
```

## 3. Figure and Axes

```

fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

## 4. Labels, Title, Legend

```

plt.plot(x, y, label="Sales")
plt.xlabel("Time")
plt.ylabel("Value")
plt.title("Sales Over Time")
plt.legend()
plt.show()
```

## 5. Line Styles and Markers

```

plt.plot(x, y, linestyle="--", marker="o")
plt.show()
```

## 6. Multiple Lines

```

y2 = [5, 15, 20, 25]

plt.plot(x, y, label="A")
plt.plot(x, y2, label="B")
plt.legend()
plt.show()
```

## 7. Subplots

```

fig, axes = plt.subplots(1, 2, figsize=(10,4))

axes[0].plot(x, y)
axes[0].set_title("Plot 1")

axes[1].plot(x, y2)
axes[1].set_title("Plot 2")

plt.show()
```

## 8. Bar Charts

```

labels = ["A", "B", "C"]
values = [10, 20, 15]

plt.bar(labels, values)
plt.show()
```

## 9. Histograms

```

data = np.random.randn(1000)
plt.hist(data, bins=30)
plt.show()
```

## 10. Scatter Plot

```

x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y)
plt.show()
```

## 11. Axis Customization

```

plt.plot([1,2,3,4], [10,20,25,30])
plt.xlim(0,5)
plt.ylim(0,35)
plt.grid(True)
plt.show()
```

## 12. Tick Labels

```

plt.plot([1,2,3,4], [10,20,25,30])
plt.xticks([1,2,3,4], ["Q1","Q2","Q3","Q4"])
plt.show()
```

## 13. Annotations

```

plt.plot([1,2,3,4], [10,20,25,30])
plt.annotate(
    "Peak",
    xy=(4, 30),
    xytext=(2, 25),
    arrowprops=dict(arrowstyle="->")
)
plt.show()
```

## 14. Saving Figures

```

plt.plot([1,2,3,4], [10,20,25,30])
plt.savefig("plot.png", dpi=300, bbox_inches="tight")
plt.show()
```

## 15. Styles

```

plt.style.use("ggplot")
plt.plot([1,2,3,4], [10,20,25,30])
plt.show()
```

## 16. Pandas Integration

```

import pandas as pd

df = pd.DataFrame({
    "year": [2019, 2020, 2021],
    "sales": [100, 150, 200]
})

df.plot(x="year", y="sales")
plt.show()
```

## 17. Time Series Plot

```

dates = pd.date_range("2023-01-01", periods=5)
values = np.random.randint(10, 50, 5)

plt.plot(dates, values)
plt.gcf().autofmt_xdate()
plt.show()
```

## 18. Log Scale

```

x = np.linspace(1, 100, 100)
y = np.exp(x / 20)

plt.plot(x, y)
plt.yscale("log")
plt.show()
```

## 19. Twin Axes

```

x = [1,2,3,4]
y1 = [10,20,25,30]
y2 = [1,2,3,4]

fig, ax1 = plt.subplots()
ax1.plot(x, y1)
ax1.set_ylabel("Sales")

ax2 = ax1.twinx()
ax2.plot(x, y2)
ax2.set_ylabel("Profit")

plt.show()
```

## 20. 3D Plot

```

from mpl_toolkits.mplot3d import Axes3D

X = np.linspace(-5,5,50)
Y = np.linspace(-5,5,50)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.plot_surface(X, Y, Z)
plt.show()
```
