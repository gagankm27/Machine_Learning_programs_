**Importing the Required Libraries**

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

**Loading IRIS dataset**

```python
data = load_iris()
X_original = data.data
y_true = data.target
```

**Fitting Standard Scalar to normalize Data**

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_original)
```

**Importing PCA model and Fitting**

```python
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X_scaled)
```

```python
print("Explained Variance Ratio by the 2 Components:", pca.explained_variance_ratio_)
```

**Importing K Means Clustering and Fitting the Model**

```python
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans.fit(X_reduced)
pca_cluster_labels = kmeans.labels_
```

**Visualization**

```python
plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_reduced[:, 0], X_reduced[:, 1],
                      c=pca_cluster_labels,
                      cmap='plasma', marker='o', s=50, alpha=0.7)
```

**Centroid Marking by K Means**

```python
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', marker='*', s=300, label='K-Means Centers')
```

**Dimentanality reduction By PCA**

```python
plt.title('K-Means Clustering on PCA-Reduced Data')
plt.xlabel('Principal Component 1 (PC1)')
plt.ylabel('Principal Component 2 (PC2)')
plt.legend()
plt.grid(True)
plt.show()
```
