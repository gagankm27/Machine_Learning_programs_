# NumPy Complete Tutorial
From Basics to Ultra-Advanced

This notebook teaches NumPy step-by-step with explanations and runnable code.

## 1. Import NumPy

```
import numpy as np
```

## 2. Creating Arrays

```

a = np.array([1, 2, 3])
b = np.array([[1, 2], [3, 4]])
a, b
```

## 3. Zeros, Ones, Ranges

```

np.zeros((2,3))
np.ones((3,3))
np.arange(0, 10, 2)
np.linspace(0, 1, 5)
```

## 4. Random Arrays

```

np.random.rand(3,2)
np.random.randn(3,2)
np.random.randint(1, 10, (3,3))
```

## 5. Array Properties

```

a = np.array([[1,2,3],[4,5,6]])
a.shape, a.ndim, a.size, a.dtype
```

## 6. Indexing and Slicing

```

a = np.array([10,20,30,40])
a[1:3]
```

## 7. 2D Indexing

```

a = np.array([[1,2,3],[4,5,6]])
a[:,1]
```

## 8. Vectorized Operations

```

a = np.array([1,2,3])
b = np.array([4,5,6])
a + b
```

## 9. Broadcasting

```

a = np.array([[1,2,3],[4,5,6]])
b = np.array([10,20,30])
a + b
```

## 10. Aggregations

```

a = np.array([[1,2],[3,4]])
np.sum(a), np.mean(a), np.std(a)
```

## 11. Reshaping

```

a = np.arange(12)
a.reshape(3,4)
```

## 12. Copy vs View

```

a = np.array([1,2,3])
b = a
c = a.copy()
a[0] = 100
b, c
```

## 13. Linear Algebra

```

A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])
A @ B
```

## 14. Determinant & Inverse

```

np.linalg.det(A)
np.linalg.inv(A)
```

## 15. Advanced Indexing

```

a = np.array([10,20,30,40])
a[[0,2]]
```

## 16. Sorting

```

a = np.array([3,1,4,2])
np.sort(a)
```

## 17. Performance: Vectorization

```

a = np.arange(5)
a += 1
a
```

## 18. Broadcasting for ML

```

X = np.random.rand(100,3)
X_norm = (X - X.mean(axis=0)) / X.std(axis=0)
X_norm[:5]
```

## 19. File Operations

```

np.save('data.npy', a)
np.load('data.npy')
```

## 20. Strides & Sliding Window

```

from numpy.lib.stride_tricks import sliding_window_view
a = np.arange(10)
sliding_window_view(a, 3)
```
